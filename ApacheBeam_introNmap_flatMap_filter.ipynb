```python
#convert each row into a list using Map
import apache_beam as beam

def stream_split(element):
  return element.split(",")

p1=beam.Pipeline()

textCollection=(
    p1
      |beam.io.ReadFromText('dept_data.txt')
      |beam.Map(stream_split)
      |beam.io.WriteToText('data/dept_data_check')
)

p1.run()
```

```python
#convert each row into a list using FlatMap
import apache_beam as beam

def stream_split(element):
  return element.split(",")

p1=beam.Pipeline()

textCollection=(
    p1
      |beam.io.ReadFromText('dept_data.txt')
      |beam.FlatMap(stream_split)#rather than converting the values in a list it
                                 # is converting it into individual elemts
      |beam.io.WriteToText('data/dept_data_check')
)

p1.run()
```

```python
#check eelemnts of only accounts dept
import apache_beam as beam

def stream_split(element):
  return element.split(",")

p1=beam.Pipeline()

textCollection=(
    p1
      |beam.io.ReadFromText('dept_data.txt')
      |beam.Map(stream_split)
      |beam.Filter(lambda x:x[3]=='Accounts')
      |beam.io.WriteToText('data/dept_data_check_accounts')
)

p1.run()
```

```python
#check attendance of only accounts dept we add 1 to the row
import apache_beam as beam

def stream_split(element):
  return element.split(",")

def add_one_to_row(record):
    record.append(1)
    return record

p1=beam.Pipeline()

textCollection=(
    p1
      |beam.io.ReadFromText('dept_data.txt')
      |beam.Map(stream_split)
      |beam.Filter(lambda x:x[3]=='Accounts')
      |beam.Map(lambda record:(record[1],1))#tuple of name and each attendance
      |beam.CombinePerKey(sum)
      |beam.io.WriteToText('data/dept_data_check_accounts')
)

p1.run()
```
